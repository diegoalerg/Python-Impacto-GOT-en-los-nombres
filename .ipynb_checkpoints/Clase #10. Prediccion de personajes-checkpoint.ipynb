```python
import pandas as pg

dg = pg.read_csv('./DATA/usa_names_data.csv')

dg.head()
```

```python
lista_nombres_usa = dg['name'].to_list()

print(lista_nombres_usa[:10])
```

```python
usa_name_year = {}

names_usa = dg['name'].to_list()

usa_years = dg['year'].to_list()

for i in range(len(names_usa)):
    usa_name_year[names_usa[i]] = usa_years[i]  #HE CREADO EL DICCIONARIO QUE VINCULA EL NOMBRE CON EL AÃ‘O
```

```python
usa_name_year_before = {}

zip_list = list(zip(names_usa, usa_years))

print(zip_list[:10])

#CREO EL DICCIONARIO

for zl in zip_list:
    if zl[0] not in usa_name_year_before and zl[1] < 2011: #ESTE ES OTRO DICCIONARIO PERO QUE ESTA ANTES DEL 2011
        usa_name_year_before[zl[0]] = zl[1]
    
```

```python
print(usa_name_year_before)
```

```python
coincidencias_despues = []

for k in casa:
    if k in usa_name_year and k not in usa_name_year_before:
         #OBTENGO LOS NOMBRES QUE COINCIDEN DESPUES DE 2011!!!!!
        coincidencias_despues.append(k)
        
print(coincidencias_despues)
        
        

        
```

```python
d = Counter(names_usa)

for lngc in coincidencias_despues:
    print(lngc,d[lngc])
```
